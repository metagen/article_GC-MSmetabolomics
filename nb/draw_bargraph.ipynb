```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 項目名
items = ['Acetic acid', 'Propionic acid', 'Butyric acid']

# 定量値
values_X = np.array([6.069272286,
2.659859134,
3.429886025])  # 条件Xの定量値
values_Y = np.array([0.552059304,
0.324048403,
0.344193521])  # 条件Yの定量値

# エラー値（標準偏差など）
errors_X = np.array([0.692148465,
0.348651664,
0.258452563])  # 条件Xのエラー値
errors_Y = np.array([0.090882383,
0.08978195,
0.203728349])  # 条件Yのエラー値 
```

```python
values_Z = np.array([8.297098906,
4.307311012,
5.121483391])  # 条件Zの定量値

errors_Z = np.array([0.620962007,
0.275833064,
0.287920935])  # 条件Zのエラー値 
```

```python
plt.rcParams["font.family"] = "Arial"
```

```python
# 項目ごとの位置を設定
x = np.arange(len(items))  # 項目数に基づいた位置の配列
width = 0.4  # 棒グラフの幅


# 棒グラフを描画
fig, ax = plt.subplots(figsize=(10, 8), dpi=500)

bars_X = ax.bar(x - width/2, values_X, width, label="With feces", yerr=errors_X, capsize=5, edgecolor="black")
bars_Y = ax.bar(x + width/2, values_Y, width, label='Without feces', yerr=errors_Y, capsize=5, edgecolor="black")

# 軸のラベル、タイトル、凡例の追加
ax.set_ylabel('Relative area', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(items, fontsize=14)
ax.legend(fontsize=14)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# 凡例の枠線を消去
ax.legend(frameon=False, fontsize=14)

# グラフを表示
plt.tight_layout()
plt.show()

fig.savefig('volatility_onlyscfa.png', dpi=500)
```

```python
# 項目ごとの位置を設定
x = np.arange(len(items))  # 項目数に基づいた位置の配列
width = 0.4  # 棒グラフの幅


# 棒グラフを描画
fig, ax = plt.subplots(figsize=(10, 8), dpi=500)

bars_X = ax.bar(x - width/2, values_X, width, label="Free + feces", yerr=errors_X, capsize=5, edgecolor="black")
bars_Y = ax.bar(x + width/2, values_Z, width, label='Salt + feces', yerr=errors_Z, capsize=5, color='green', edgecolor="black")

# 軸のラベル、タイトル、凡例の追加
ax.set_ylabel('Relative area', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(items, fontsize=14)
ax.legend(fontsize=14)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# 凡例の枠線を消去
ax.legend(frameon=False, fontsize=14)

# グラフを表示
plt.tight_layout()
plt.show()

fig.savefig('freevssalt_onlyscfa.png', dpi=500)
```
